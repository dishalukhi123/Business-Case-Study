```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde


```

```python
math_scores = [65, 70, 75, 80, 85, 90, 95, 88, 78, 72]
physics_scores = [60, 68, 74, 79, 83, 89, 94, 85, 76, 70]

df = pd.DataFrame({
    "Math": math_scores,
    "Physics": physics_scores
})

df
```

```python
covariance = df.cov()
covariance
```

```python
pearson_corr = df.corr(method='pearson')
pearson_corr
```

```python
shuffled_physics = np.random.permutation(physics_scores)

df_shuffled = pd.DataFrame({
    "Math": math_scores,
    "Physics_Shuffled": shuffled_physics
})

spearman_corr = df_shuffled.corr(method='spearman')
spearman_corr
```

```python
uniform_data = np.random.uniform(0, 1, 10000)

plt.hist(uniform_data, bins=50, edgecolor='black', alpha=0.7, color='orange', density=True)
plt.title("Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
binomial_data = np.random.binomial(n=20, p=0.5, size=10000)

normal_data = np.random.normal(
    loc=np.mean(binomial_data),
    scale=np.std(binomial_data),
    size=10000
)

plt.hist(binomial_data, bins=30, alpha=0.7, label="Binomial", color='red')
plt.hist(normal_data, bins=30, alpha=0.7, label="Normal", color='green')
plt.legend()
plt.title("Binomial vs Normal Distribution")
plt.show()
```

```python
poisson_data = np.random.poisson(lam=5, size=10000)

plt.hist(poisson_data, bins=30, color='purple', alpha=0.7)
plt.title("Poisson Distribution (Daily Customers)")
plt.xlabel("Customers per Day")
plt.ylabel("Frequency")
plt.show()
```

```python
population = np.random.exponential(scale=2, size=100000)

sample_means = [
    np.mean(np.random.choice(population, size=30))
    for _ in range(1000)
]

plt.hist(sample_means, bins=30 , alpha=0.7, color='pink', edgecolor='black' , density=True)
kde = gaussian_kde(sample_means)
x_vals = np.linspace(min(sample_means), max(sample_means), 1000)
plt.plot(x_vals, kde(x_vals))
plt.title("Sample Means (n = 30)")
plt.xlabel("Mean Value")
plt.ylabel("Frequency")
plt.show()
```

```python
sample_means = [
    np.mean(np.random.choice(population, size=100))
    for _ in range(1000)
]

plt.hist(sample_means, bins=30, alpha=0.7, color='cyan', edgecolor='black')
plt.title("Sample Means (n = 100)")
plt.xlabel("Mean Value")
plt.ylabel("Frequency")
plt.show()
```
