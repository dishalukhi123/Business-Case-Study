```python
import numpy as np
from scipy.stats import norm
from scipy import stats

```

```python
population_mean = 5            # population mean
population_standard_deviation = 0.50      # population standard deviation
sample_size = 40            # sample size
sample_mean = 4.8       # sample mean
alpha = 0.02      # significance level
```

```python
z_value = (sample_mean - population_mean) / (population_standard_deviation / np.sqrt(sample_size))
z_value
```

```python
z_critical = norm.ppf(alpha)
z_critical
```

```python
if z_value <= z_critical:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
population_mean = 4 # population mean
sample_mean = 3.5 # sample mean
population_standard_deviation = 1.2 # sample standard deviation
sample_size = 25 # sample size
```

```python
t_value = (sample_mean - population_mean) / (population_standard_deviation / np.sqrt(sample_size))
t_value
```

```python
p_value = 2 * (1 - stats.t.cdf(abs(t_value), df=sample_size-1))
p_value
```

```python
sem = population_standard_deviation / np.sqrt(sample_size)

ci = stats.t.interval(
    0.95,
    df=sample_size-1,
    loc=sample_mean,
    scale=sem
)

ci
```
