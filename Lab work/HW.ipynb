```python
import numpy as np
from scipy.stats import norm
```

```python
mu = 5            # population mean
sigma = 0.50      # population standard deviation
n = 40            # sample size
x_bar = 4.8       # sample mean
alpha = 0.02      # significance level
```

```python
z_value = (x_bar - mu) / (sigma / np.sqrt(n))
z_value
```

```python
z_critical = norm.ppf(alpha)
z_critical
```

```python
if z_value <= z_critical:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
