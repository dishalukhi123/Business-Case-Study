### Descriptive Statistics
Descriptive statistics summarize and organize data to describe its main features.
They do not make predictions, only explain what the data shows.

Examples:
- Mean
- Median
- Mode
- Variance
- Standard Deviation


### Inferential Statistics
Inferential statistics use sample data to make predictions or conclusions about a population.

Examples:
- Hypothesis testing
- Confidence intervals
- Regression analysis
- Probability distributions


| Descriptive Statistics | Inferential Statistics |
|-----------------------|------------------------|
| Describes data        | Predicts outcomes      |
| Uses entire dataset   | Uses samples           |
| No assumptions        | Uses probability       |
| Mean, median, charts | Hypothesis testing     |


```python
import pandas as pd

data = {
    "Student": ["A", "B", "C", "D", "E"],
    "Marks": [65, 70, 75, 80, 85]
}

df = pd.DataFrame(data)
df
```

```python
df.describe()
```

- **count**: Number of non-null values
- **mean**: Average value
- **std**: Standard deviation (spread of data)
- **min**: Smallest value
- **25%**: First quartile
- **50%**: Median
- **75%**: Third quartile
- **max**: Largest value


```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
df["Marks"].mean()
np.mean(df["Marks"])
```

```python
df["Marks"].median()
np.median(df["Marks"])
```

```python
df["Marks"].mode()
```

```python
values = {
    "Mean": df["Marks"].mean(),
    "Median": df["Marks"].median(),
    "Mode": df["Marks"].mode()[0]
}

plt.bar(values.keys(), values.values())
plt.title("Comparison of Mean, Median, and Mode")
plt.ylabel("Marks")
plt.show()
```

```python
range_value = df["Marks"].max() - df["Marks"].min()
range_value
```

```python
df["Marks"].var()
np.var(df["Marks"], ddof=1)
```

```python
df["Marks"].std()
np.std(df["Marks"])
```

```python
plt.boxplot(df["Marks"])
plt.title("Boxplot of Marks")
plt.ylabel("Marks")
plt.show()
```

The boxplot shows moderate dispersion with no extreme outliers.
The data is evenly spread around the median.
