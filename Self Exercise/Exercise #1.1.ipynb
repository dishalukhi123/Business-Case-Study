## Q.2 How is Statistics applied in Data Science?
**Answer:**  
In Data Science, statistics is used to:
- Summarize data using descriptive statistics
- Identify patterns and trends
- Perform hypothesis testing
- Build predictive models
- Measure uncertainty and accuracy
- Statistics forms the foundation for machine learning and data-driven decisions.

## Q.3 What is the difference between Data and Information in the context of Statistics?

**Answer:**  
- Data: Raw facts and figures without meaning (e.g., exam scores).
- Information: Processed and organized data that provides meaning (e.g., average score, pass percentage).

### Q.6 Differentiate between Numerical Data and Categorical Data with examples.

**Answer:**  

| Numerical Data | Categorical Data |
|--------------|----------------|
| Measurable values | Category-based values |
| Represented by numbers | Represented by labels |
| Example: Age, Height | Example: Gender, Blood Group |

### Q.7 What is the difference between Discrete and Continuous numerical data?

**Answer:**  
- **Discrete Data:** Countable values (e.g., number of students, number of cars).  
- **Continuous Data:** Measurable values that can take decimals (e.g., weight, temperature).

### Q.9 Explain the use of bar graphs, line charts, and histograms in data representation.

**Answer:**  
- **Bar Graphs:** Used to compare different categories.  
- **Line Charts:** Used to show trends over time.  
- **Histograms:** Used to show the distribution of numerical data.

---

### Q.10 How can pie charts and box plots help in understanding the distribution of data?

**Answer:**  
- **Pie Charts:** Show the proportion or percentage of each category.  
- **Box Plots:** Show data distribution, median, quartiles, and outliers.

---

### Q.11 What is the difference between Population and Sample in statistics?

**Answer:**  
- **Population:** The entire group under study.  
- **Sample:** A subset of the population selected for analysis.

---

### Q.12 Why do data scientists often work with samples instead of entire populations?

**Answer:**  
Data scientists work with samples because studying the entire population is time-consuming, costly, and often impractical. Samples provide reliable results with fewer resources.
